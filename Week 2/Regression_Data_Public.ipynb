```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
# memberikan path file dataset
path = '/content/clubs.csv'

# Membaca dataset menggunakan pandas
df = pd.read_csv(path)

# menampilkan beberapa baris pertama dari dataset
df.head()
```

```python
#memberikan informasi dataset
df.info()
```

```python
#memberikan statistik deskriptif kolom yang ada pada dataset
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat plot garis
sns.lineplot(data=df, x="club_id", y="squad_size")

# Menambahkan judul plot
plt.title('club id vs. squad size')

# Menambahkan label sumbu x dan y
plt.xlabel('club_id')
plt.ylabel('squad_size')

# Menampilkan plot
plt.show()
```

```python
X = df[['club_id']]
y = df['squad_size']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print(f'Koefisien: {model.coef_[0]}') # mencetak koefisien model regresi
print(f'Intercept: {model.intercept_}') # mencetak nilai intercept
print(f'Skor R-squared: {model.score(X, y)}') # mencetak skor R-squared dari model
```
